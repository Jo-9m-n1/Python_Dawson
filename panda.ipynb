```python
import pandas as pd

data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data)
df
```

```python
print(df.loc[0]) # prints the 0th row
print(df.loc[[0, 1]]) # prints the 0th and 1st row
```

```python
import pandas as pd

data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
df
```

```python
print(df.loc["day1"]) # prints the row with index "day1"
```
