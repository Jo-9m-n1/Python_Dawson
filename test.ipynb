```python
import pandas as pd
import matplotlib.pyplot as plt

def euler(y, m, a, c, p, g, dt=0.01):
    print(f"y: {y}, m: {m}, a: {a}, c: {c}, p: {p}, g: {g}, dt: {dt}")

    ts = []
    ys = []

    t = 0
    vy = 0

    while y > 0:
        ac = -g + ((c * p * a) / (2 * m)) * (vy**2)
        vy += ac * dt
        y += vy * dt
        t += dt

        ts += [round(t, 3)]
        ys += [round(y, 3)]

    print("t , y")
    for t, y in zip(ts, ys):
        print(t, y)

    plt.plot(ts, ys)
    plt.title("Position vs Time")
    plt.xlabel("Time (s)")
    plt.ylabel("Position (m)")
    plt.show()

euler(1.55, 0.89 / 1000, 0.01339, 0.9824, 1.204, 9.81, 0.1)
```
